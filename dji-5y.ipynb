```python
import pandas
import datetime
```

```python
debt = pandas.read_csv("stock-5y.csv", header= None)
debt
```

```python
dji = pandas.read_csv("DJI-5y.csv")
dji
```

```python
import matplotlib.pyplot as plt
dji['Close'].plot()
```

```python
debt[1].plot()
```

```python
# Find change in value of debt for each row
debt_change = debt.set_index(0).diff()
debt_change.reset_index(level=0, inplace=True)
debt_change
```

```python
pos_debt_change = debt_change.loc[debt_change[1] <0]
pos_debt_change.columns = ['Date','Change']
pos_debt_change['Date'] = pos_debt_change['Date'].apply(lambda x: pandas.to_datetime(x)+datetime.timedelta(days=8))
pos_debt_change
```

```python

sell_dates = pos_debt_change['Date']
sell_dates
```

```python
# Find change in value of DJI for each row
dji_change = dji.set_index('Date').pct_change()
dji_change.reset_index(level=0, inplace=True)
dji_change
```

```python
traded_dji_change = dji_change[pandas.to_datetime(dji_change['Date']).isin(sell_dates)]
traded_dji_change['Date'] = traded_dji_change['Date'].apply(lambda x: pandas.to_datetime(x))
traded_dji_change = traded_dji_change.merge(pos_debt_change, on='Date')
traded_dji_change['profit_loss'] = traded_dji_change['Close']*(traded_dji_change['Change'].abs())
traded_dji_change['profit_loss_cumsum'] = traded_dji_change.profit_loss.cumsum()
traded_dji_change
```

```python
traded_dji_change['profit_loss_cumsum'].plot()
```

```python
valid_dji_change = dji_change[1:]
valid_dji_change['profit_loss_cumsum'] = valid_dji_change.Close.cumsum()
valid_dji_change
```

```python
valid_dji_change['profit_loss_cumsum'].plot()
```
