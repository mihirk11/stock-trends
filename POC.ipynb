```python
import pandas
```

```python
debt = pandas.read_csv("stock.csv", header= None)
debt
```

```python
dji = pandas.read_csv("DJI.csv")
dji
```

```python
import matplotlib.pyplot as plt
dji['Close'].plot()
```

```python
debt[1].plot()
```

```python
# Find change in value of debt for each row
debt_change = debt.set_index(0).diff()
debt_change.reset_index(level=0, inplace=True)
debt_change
```

```python
pos_debt_change = debt_change.loc[debt_change[1] >0]
pos_debt_change
```

```python
import datetime
sell_dates = pandas.Series([pandas.to_datetime(date)+datetime.timedelta(days=8) for date in pos_debt_change[0]])
sell_dates
```

```python
# Find change in value of DJI for each row
dji_change = dji.set_index('Date').pct_change()
dji_change.reset_index(level=0, inplace=True)
dji_change
```

```python
traded_dji_change = dji_change[pandas.to_datetime(dji_change['Date']).isin(sell_dates)]
traded_dji_change['Close_cumsum'] = traded_dji_change.Close.cumsum()
traded_dji_change
```

```python
traded_dji_change['Close_cumsum'].plot()
```

```python
valid_dji_change = dji_change[1:]
valid_dji_change['Close_cumsum'] = valid_dji_change.Close.cumsum()
valid_dji_change
```

```python
valid_dji_change['Close_cumsum'].plot()
```
